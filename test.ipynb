```python
import os
from dotenv import load_dotenv
load_dotenv()

# NOTE: uses modern client
from openai import OpenAI

print("openai package version:", __import__("openai").__version__)

key = os.getenv("OPENAI_API_KEY")
if not key:
    print("OPENAI_API_KEY not set in env. Set it before running.")
else:
    client = OpenAI(api_key=key)
    print("Client OK (no network call).")
```
